```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from scipy.stats import ttest_ind

pd.set_option('display.max_rows', 50) # показывать больше строк
pd.set_option('display.max_columns', 50) # показывать больше колонок

study = pd.read_csv('stud_math.csv')
```

# Функции

```python
def get_boxplot(column): # Распределение баллов в зависимости от переменных с помощью box-plot. Distribution of points depending on variables using box-plot
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=study.loc[study.loc[:, column].isin(study.loc[:, column].value_counts().index[:10])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
    
def get_stat_dif(column):
    cols = study.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        ttest = ttest_ind(
            study.loc[study.loc[:, column] == comb[0], 'score'].dropna(),
            study.loc[study.loc[:, column] == comb[1], 'score'].dropna()).pvalue
        #         print(f"для столбца {column} ttest:= {ttest}")
        #         print(f"                     comb:= {combinations_all}")
        if ttest <= 0.05 / len(combinations_all):  # Учли поправку Бонферони
            print('!!!Найдены статистически значимые различия для колонки',
                  column)
            break
            
```

# Absences

absences — количество пропущенных занятий

```python
print(
    f'Количество пропущенных значений в столбце absences = {study.absences.isnull().sum()}'
)
study.absences.hist(color='blue', bins=70)
study.absences.describe()
plt.title('absences')
```

Есть несколько выбросов, которых не могло быть (212 и 385), так как учебных дней в году меньше. Воспользуемся формулой интеркватрального расстояния.

```python
print(
    f'Количество пропущенных значений в столбце absences = {study.absences.isnull().sum()}'
)
median = study.absences.median()
IQR = study.absences.quantile(0.75) - study.absences.quantile(0.25)
perc25 = study.absences.quantile(0.25)
perc75 = study.absences.quantile(0.75)
print(
    '25-й перцентиль: {},'.format(perc25),
    '75-й перцентиль: {},'.format(perc75), "IQR: {}, ".format(IQR),
    "Границы выбросов: [{f}, {l}].".format(f=perc25 - 1.5 * IQR,
                                           l=perc75 + 1.5 * IQR))
```

Проверим что за данные мы отфильтровали.

```python
study.loc[~study.absences.between(perc25 - 1.5*IQR, perc75 + 1.5*IQR), 'absences']
```

Здесь довольно мого данных до 28 часов пропусков. Оставим их тоже. Этот способ позволил отобрать значения пропусков экстремально большие. Отфильтруем эти данные.

```python
study = study.loc[study.absences.between(perc25 - 1.5*IQR, 28)|pd.isnull(study.absences)]
```

```python
study.absences.loc[study.absences.between(perc25 - 1.5 * IQR, 28)].hist(bins=30,
                                                         range=(0, 30),
                                                         label='IQR',
                                                         color='blue')
plt.title('absences')
```

# Score

score — баллы по госэкзамену по математике

```python
print(
    f'Количество пропущенных значений в столбце absences = {study.score.isnull().sum()}'
)
study.score.hist(color='blue', bins=70)
study.score.describe()
plt.title('score')
```

Это наша целевая переменная. Распределение нормальное, если не считать учеников, получивших 0 баллов. Пропусков всего 6, заменим их на среднее значение

Нулевая оценка, по моему мнению, получена в результате нарушения, которое может быть не связана с успеваемостью. Например, не пришел на экзамен. Потому. Создадим еще один столбец 'no_score' в котором "1" присвоим нулевому баллу на экзамене, и "0", если оценка отличается от нулевой. Score - целевая переменная, поэтому менять ничего не будем.

```python
study['no_score'] = study.score.apply(lambda x: 1 if x == 0 else 0)
study.no_score.value_counts()
```

Имеет смысл провести тест Стюдента по параметру no_score - получившие или не получившие оценки.

```python
def get_stat_dif(column):
    cols = study.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        ttest = ttest_ind(
            study.loc[study.loc[:, column] == comb[0], 'no_score'].dropna(),
            study.loc[study.loc[:, column] == comb[1],
                     'no_score'].dropna()).pvalue
        #         print(f"для столбца {column} ttest:= {ttest}")
        #         print(f"                     comb:= {combinations_all}")
        if ttest <= 0.05 / len(combinations_all):  # Учли поправку Бонферони
            print('!!!Найдены статистически значимые различия для колонки',
                  column)
            break
```
